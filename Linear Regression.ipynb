### Linear Regression from Scratch

```python
import numpy as np

```

#### Variable intialization with arbitrary slope and intercept

```python
np.random.seed(0)
m = 3
c = 4
x = np.arange(0,50)
x =x.reshape(1,len(x))
N = x.shape[1]
y = m*x + c + 50*np.random.rand(1,N)
```

```python
x.shape
```

### Data

```python
import matplotlib.pyplot as plt
plt.plot(x,y,'go')

plt.ylabel('Data')
plt.show()
```

### Learning the slope and intercept by minimizing the error

```python
import matplotlib.pyplot as plt

m1 = np.random.randn(1,1)
c1 = np.random.randn(1,1)
learningrate = 0.001
numiter = 50
for i in range(numiter):
    dm = (2.0/N) *  -np.sum(np.multiply(x,(y - (m1*x +c1))))
   
    dc = (2.0/N) * -np.sum(y - (m1*x + c1))
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dc
    # print ("no of iterations:",i )
    # plt.plot(x,y,'go')
    # plt.plot(x,m1*x+c1,'r*')
    # plt.ylabel('Data')
    # plt.show()
plt.plot(x,y,'go')
plt.plot(x,m1*x+c1,'r*')
plt.ylabel('Data')
plt.show()
    
```
